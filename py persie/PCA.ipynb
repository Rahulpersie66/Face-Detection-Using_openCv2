```python
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
cancer.keys()
```

```python
#real time datamust need to be normalized(via standard scaling)

df=pd.DataFrame(cancer['data'],columns= cancer ['feature_names'])
```

```python
df.head()
```

```python
#Scaling standard deviation
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
```

```python
# transform the df
scaled_data=scaler.transform(df)
scaled_data
```

```python
from sklearn.decomposition import PCA
pca =PCA(n_components=2)#how much component or matrix you want to reduce i.e 2*2 matrix
```

```python
pca.fit(scaled_data)
```

```python
x_pca=pca.transform(scaled_data)
```

```python
scaled_data.shape#shows no. of columns and rows inmatrix
```

```python
x_pca.shape#no reduced 30 rows to2 rows , this is pca reduced higher dimension to lower dimension
```

```python
#want to see allcomponents of PCA
#pca.components
#plot the graph the cmponents you obtained
plt.figure(figsize =(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='rainbow')#this changes the color wit cmap -=rainbow
plt.xlabel("First PCA")
plt.ylabel("Second PCA")
plt.show()
```
